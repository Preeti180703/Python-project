```python
# which subject pair has the strongest correlation?
corr_matrix=data_student[['sql_marks','excel_marks','python_marks','power_bi_marks','english_marks']].corr()
print('correlation matrix:\n',corr_matrix)
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title('subject correlation heatmap')
plt.show()
```

```python
# Student weak in english but strong in technical.
weak_eng=data_student[(data_student[['sql_marks','excel_marks','python_marks',
                                    'power_bi_marks']].mean(axis=1)>75) & (data_student['english_marks']<71)]
print("students good in technical but weak in english:")
print(weak_eng[['student_id','location','english_marks']])
```

```python
# Count of students by grade category
def grade(avg):
    if avg>=80:
        return 'A'
    elif avg>=60:
        return 'B'
    elif avg>=40:
        return 'C'
    else:
        return 'Fail'
data_student['avg_marks']=data_student[['sql_marks','excel_marks','python_marks',
                                        'power_bi_marks','english_marks']].mean(axis=1)
data_student['grade']=data_student['avg_marks'].apply(grade)
data_student['grade'].value_counts()
```
